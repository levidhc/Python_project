```python
import pandas as pd
import numpy as np
```

```python
screener = []
def add_filter(filter, min=None, max=None):
    add = {'filter':filter, 'min':min, 'max':max}
    screener.append(add)
    return screener
```

```python
add_filter('POR', 10, 20)
add_filter('PER', 20, 30)
```

```python
a = pd.read_pickle('screener.pkl')
```

```python
a[screener[0]['filter']][a[screener[0]['filter']].between(screener[0]['min'], screener[0]['max'])]
```

```python
filtered = None
for i in range(len(screener)):
    if filtered is None:
        a.index.name = 'code'
        filtered = a[screener[i]['filter']][a[screener[i]['filter']].between(screener[i]['min'],screener[i]['max'])].reset_index().set_index('code')
    else:
        filtered = filtered.join(a[screener[i]['filter']][a[screener[i]['filter']].between(screener[i]['min'],screener[i]['max'])], how='inner')
```
